```python
def missing_data_delete(data):
    total = data.isnull().sum().sort_values(ascending = False)
    percent = (data.isnull().sum()/data.isnull().count()*100).sort_values(ascending = False)
    df = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
    dfx=df[df['Percent']>=10]
    
    column_list=list(dfx.index.values)
    print(column_list)
    
    df_data=data.drop(column_list, axis=1)
    return df_data

df_fin=missing_data_delete(application_test)
df_fin.head(10)
```
